```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
import glob
import shutil
```

```python
# Configuração de Data Augmentation
datagen = ImageDataGenerator(
    rotation_range=40,        # Rotação aleatória até 40 graus
    width_shift_range=0.2,    # Deslocamento horizontal de até 20%
    height_shift_range=0.2,   # Deslocamento vertical de até 20%
    shear_range=0.2,          # Cisalhamento
    zoom_range=0.2,           # Zoom in e out
    horizontal_flip=True,     # Flip horizontal
    fill_mode='nearest'       # Preenchimento de áreas vazias
)
```

```python
# Diretório de entrada e saída
input_dir = "/content/"  # Substitua pelo caminho do diretório com suas imagens
output_dir = "augmented_images"
os.makedirs(output_dir, exist_ok=True)  # Cria o diretório de saída, se não existir
```

```python
# Encontrar todas as imagens no diretório de entrada
image_paths = glob.glob(os.path.join(input_dir, "*.jpg"))  # Ajuste o padrão se necessário
```

```python
# Iterar sobre cada imagem no diretório
for image_path in image_paths:
    # Carregar a imagem
    original_image = Image.open(image_path)

    # Converter para array numpy
    image_array = np.array(original_image)
    image_array = np.expand_dims(image_array, axis=0)  # Adicionar dimensão do batch

    # Gerar imagens aumentadas
    generated_images = datagen.flow(image_array, batch_size=1)

    # Nome base da imagem original
    base_name = os.path.splitext(os.path.basename(image_path))[0]

    # Gerar e salvar 10 imagens aumentadas
    for i in range(10):
        batch = next(generated_images)  # Obter próxima imagem gerada
        augmented_image = batch[0].astype('uint8')  # Converter para uint8

        # Criar o caminho para salvar a imagem
        output_path = os.path.join(output_dir, f"{base_name}_augmented_{i+1}.jpg")

        # Salvar a imagem aumentada
        Image.fromarray(augmented_image).save(output_path)
        print(f"Imagem gerada e salva: {output_path}")

    # Compactar a pasta de imagens aumentadas
    zip_file = "augmented_images.zip"
    shutil.make_archive("augmented_images", 'zip', output_dir)
    print(f"Pasta compactada em: {zip_file}")

    # Fornecer a opção de download (se estiver rodando em um ambiente como Colab)
    try:
      from google.colab import files
      files.download(zip_file)
      print("Arquivo disponível para download.")
    except ImportError:
      print("Download automático disponível apenas no Google Colab. O arquivo ZIP foi gerado.")
```

```python
# Exibindo as imagens geradas
fig, axes = plt.subplots(1, 5, figsize=(20, 10))  # Mostra 5 imagens
for i in range(5):
    img = Image.open(os.path.join(output_dir, f"augmented_image_{i+1}.jpg"))
    axes[i].imshow(img)
    axes[i].axis("off")
plt.show()
```
