```python
# -*- coding: utf-8 -*-
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
import time

# --- 1. Partial Cross-Entropy Loss ---
# (Remains the same)
def partial_categorical_crossentropy(y_true, y_pred, ignore_label=0):
    y_true = tf.cast(y_true, tf.int32)
    if len(y_true.shape) == 4 and y_true.shape[-1] == 1:
        y_true = tf.squeeze(y_true, axis=-1)
    mask = tf.not_equal(y_true, ignore_label)
    mask = tf.cast(mask, tf.float32)
    num_classes = tf.shape(y_pred)[-1]
    y_true_one_hot = tf.one_hot(y_true, depth=num_classes, dtype=tf.float32)
    loss_pixelwise = tf.keras.losses.categorical_crossentropy(
        y_true_one_hot, y_pred, from_logits=True)
    masked_loss = loss_pixelwise * mask
    total_loss = tf.reduce_sum(masked_loss)
    total_valid_pixels = tf.reduce_sum(mask) + 1e-8
    mean_loss = tf.math.divide_no_nan(total_loss, total_valid_pixels)
    return mean_loss

# --- 2. Model Definition with Overfitting Mitigation ---

L2_REG = 1e-5 # Regularization factor

def conv_block(input_tensor, num_filters, kernel_size=(3, 3), add_dropout=False, dropout_rate=0.2):
    """Conv block with optional Dropout and L2 regularization."""
    x = layers.Conv2D(num_filters, kernel_size, padding="same",
                      kernel_regularizer=keras.regularizers.l2(L2_REG))(input_tensor)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    # Optional Dropout after first activation
    if add_dropout:
        x = layers.Dropout(dropout_rate)(x)
    x = layers.Conv2D(num_filters, kernel_size, padding="same",
                      kernel_regularizer=keras.regularizers.l2(L2_REG))(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    return x

def decoder_block(input_tensor, skip_tensor, num_filters, add_dropout=False, dropout_rate=0.25):
    """Decoder block with optional Dropout."""
    x = layers.Conv2DTranspose(num_filters, (2, 2), strides=2, padding="same",
                               kernel_regularizer=keras.regularizers.l2(L2_REG))(input_tensor)
    x = layers.Concatenate()([x, skip_tensor])
    # Add dropout within the conv_block of the decoder
    x = conv_block(x, num_filters, add_dropout=add_dropout, dropout_rate=dropout_rate)
    return x

def build_unet_with_backbone(input_shape, num_classes, backbone_name="MobileNetV2"):
    """Builds U-Net with backbone, Dropout, and L2 Regularization."""
    if backbone_name == "MobileNetV2":
        backbone = keras.applications.MobileNetV2(
            input_shape=input_shape, include_top=False, weights='imagenet'
        )
        skip_connection_names = [
            "block_1_expand_relu", "block_3_expand_relu",
            "block_6_expand_relu", "block_13_expand_relu"
        ]
        bottleneck_output_name = "out_relu"
    else:
        raise ValueError(f"Unsupported backbone: {backbone_name}")

    # Keep backbone trainable for adaptation (can be frozen initially for transfer learning)
    # backbone.trainable = False

    inputs = backbone.input
    skip_outputs = [backbone.get_layer(name).output for name in skip_connection_names]
    bottleneck = backbone.get_layer(bottleneck_output_name).output

    # Decoder with Dropout added
    d1 = decoder_block(bottleneck, skip_outputs[3], 512, add_dropout=True, dropout_rate=0.3)
    d2 = decoder_block(d1, skip_outputs[2], 256, add_dropout=True, dropout_rate=0.25)
    d3 = decoder_block(d2, skip_outputs[1], 128, add_dropout=True, dropout_rate=0.2)
    d4 = decoder_block(d3, skip_outputs[0], 64, add_dropout=False) # Less dropout near output

    final_upsample = layers.Conv2DTranspose(32, (2, 2), strides=2, padding="same",
                                           kernel_regularizer=keras.regularizers.l2(L2_REG))(d4)
    final_conv = conv_block(final_upsample, 32, kernel_size=(1,1))

    # Output layer - No activation if using from_logits=True in loss
    outputs = layers.Conv2D(num_classes, (1, 1), activation=None, name="segmentation_output")(final_conv)
    model = keras.Model(inputs=inputs, outputs=outputs, name=f"UNet_{backbone_name}_Regularized")
    return model

# --- 3. Configuration and Data Simulation ---
IMG_HEIGHT = 256
IMG_WIDTH = 256
IMG_CHANNELS = 3
NUM_CLASSES = 5
IGNORE_LABEL = 0
INPUT_SHAPE = (IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS)
EPOCHS = 50 # Increase epochs since we have early stopping
BATCH_SIZE = 4
EARLY_STOPPING_PATIENCE = 5 # Stop after 5 epochs with no improvement in val_loss

# Build the regularized model
model = build_unet_with_backbone(INPUT_SHAPE, NUM_CLASSES, backbone_name="MobileNetV2")
model.summary() # Print model summary

# Simulate Data (same as before)
print("Simulating dummy data for demonstration...")
NUM_TRAIN_SAMPLES = 20; NUM_VAL_SAMPLES = 5
x_train = np.random.rand(NUM_TRAIN_SAMPLES, IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS).astype(np.float32)
x_val = np.random.rand(NUM_VAL_SAMPLES, IMG_HEIGHT, IMG_WIDTH, IMG_CHANNELS).astype(np.float32)
y_train_partial = np.full((NUM_TRAIN_SAMPLES, IMG_HEIGHT, IMG_WIDTH), IGNORE_LABEL, dtype=np.int32)
y_val_partial = np.full((NUM_VAL_SAMPLES, IMG_HEIGHT, IMG_WIDTH), IGNORE_LABEL, dtype=np.int32)
for i in range(NUM_TRAIN_SAMPLES):
    num_points = np.random.randint(50, 200); class_ids = np.random.randint(1, NUM_CLASSES, size=num_points)
    rows, cols = np.random.randint(0, IMG_HEIGHT, size=num_points), np.random.randint(0, IMG_WIDTH, size=num_points)
    y_train_partial[i, rows, cols] = class_ids
for i in range(NUM_VAL_SAMPLES):
    num_points = np.random.randint(50, 200); class_ids = np.random.randint(1, NUM_CLASSES, size=num_points)
    rows, cols = np.random.randint(0, IMG_HEIGHT, size=num_points), np.random.randint(0, IMG_WIDTH, size=num_points)
    y_val_partial[i, rows, cols] = class_ids
y_val_full = np.random.randint(1, NUM_CLASSES, size=(NUM_VAL_SAMPLES, IMG_HEIGHT, IMG_WIDTH), dtype=np.int32)
print(f"Created dummy data: {NUM_TRAIN_SAMPLES} train, {NUM_VAL_SAMPLES} validation samples.")
print(f"Training will run for up to {EPOCHS} epochs with Early Stopping (patience={EARLY_STOPPING_PATIENCE}).")


# --- 4. Compile and Train the Model (Simulated) ---

# Compile only with the loss function
model.compile(optimizer=keras.optimizers.Adam(learning_rate=1e-4),
              loss=partial_categorical_crossentropy)

# Add EarlyStopping callback
early_stopping_callback = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss', # Monitor validation loss
    patience=EARLY_STOPPING_PATIENCE,
    verbose=1,
    mode='min', # Stop when the monitored quantity stops decreasing
    restore_best_weights=True # Restore weights from the epoch with the best val_loss
)

print(f"\n--- Starting Simulated Training (Max {EPOCHS} Epochs) ---")

try:
    history = model.fit(
        x_train,
        y_train_partial,
        batch_size=BATCH_SIZE,
        epochs=EPOCHS,
        callbacks=[early_stopping_callback], # Add the callback here
        validation_data=(x_val, y_val_partial),
        verbose=1
    )
    print("\n--- Simulated Training Finished ---")
    training_successful = True

    # Plot training history (loss only)
    plt.figure(figsize=(6, 4))
    plt.plot(history.history['loss'], label='Training Loss (Partial)')
    plt.plot(history.history['val_loss'], label='Validation Loss (Partial)')
    plt.title('Training and Validation Loss (Partial Labels)')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"\n--- ERROR during Training: ---"); print(e); print("-------------------------------")
    training_successful = False

# --- Helper function for manual MeanIoU calculation ---
def calculate_miou_from_logits(y_true_full, y_pred_logits, num_classes, ignore_label=None):
    """Calculates MeanIoU from integer labels and prediction logits."""
    # Get predicted class IDs from logits
    y_pred_labels = tf.argmax(y_pred_logits, axis=-1, output_type=tf.int32) # Shape (B, H, W)
    y_true_full = tf.cast(y_true_full, tf.int32) # Ensure y_true is int32

    # Flatten predictions and ground truth
    y_pred_flat = tf.reshape(y_pred_labels, [-1])
    y_true_flat = tf.reshape(y_true_full, [-1])

    # Optional: Filter out ignore_label if provided
    if ignore_label is not None:
        valid_mask = tf.not_equal(y_true_flat, ignore_label)
        y_pred_flat = tf.boolean_mask(y_pred_flat, valid_mask)
        y_true_flat = tf.boolean_mask(y_true_flat, valid_mask)

    # Calculate confusion matrix
    # Use weights=None since we already filtered ignored labels
    confusion_matrix = tf.math.confusion_matrix(
        y_true_flat, y_pred_flat, num_classes=num_classes, dtype=tf.float32
    )

    # Calculate IoU per class
    true_positives = tf.linalg.tensor_diag_part(confusion_matrix)
    predicted_positives = tf.reduce_sum(confusion_matrix, axis=0)
    actual_positives = tf.reduce_sum(confusion_matrix, axis=1)

    intersection = true_positives
    union = predicted_positives + actual_positives - intersection

    # Avoid division by zero for classes with zero union
    # If union is 0, IoU is NaN. We convert NaN to 0.
    iou = tf.math.divide_no_nan(intersection, union)

    # Calculate MeanIoU (average over classes that had actual positives)
    # Or simply average over all non-ignore classes depending on definition needed
    # Here, we average over all classes present in the confusion matrix dimension (excluding ignore if filtered)
    # If ignore_label was filtered *before* confusion matrix, num_classes might need adjustment,
    # but confusion_matrix should handle the correct range if num_classes is correct.
    # Let's average the calculated IoUs, excluding NaNs which become 0.
    valid_iou = tf.boolean_mask(iou, tf.math.is_finite(iou)) # Filter out potential NaNs if needed, though divide_no_nan helps
    mean_iou = tf.reduce_mean(valid_iou) if tf.size(valid_iou) > 0 else 0.0

    return mean_iou.numpy()


# --- 5. Evaluate the Model Manually (with Full Masks) ---
if training_successful:
    print("\n--- Evaluating Model Performance Manually (with Full Masks) ---")

    print("Generating predictions on FULL validation masks...")
    # Predict on the validation set (using full masks for evaluation)
    val_predictions_logits = model.predict(x_val, batch_size=BATCH_SIZE)

    print("Calculating MeanIoU manually...")
    # Calculate MeanIoU using the helper function
    mean_iou_manual = calculate_miou_from_logits(
        y_val_full, val_predictions_logits, NUM_CLASSES, ignore_label=IGNORE_LABEL
    )

    print("\nManual Evaluation Metrics on FULL Validation Masks (Dummy Data):")
    print(f"  - Mean IoU (manual): {mean_iou_manual:.4f}")
    print("(Note: Loss is not recalculated here, focus is on interpretable metrics like mIoU)")

# --- 6. Request Input, Make Prediction, and Discuss Metrics (English) ---
# (Load/preprocess image function remains the same)
def load_and_preprocess_image(image_path, target_size=(IMG_HEIGHT, IMG_WIDTH)):
    try:
        img = tf.io.read_file(image_path)
        img = tf.io.decode_image(img, channels=IMG_CHANNELS, expand_animations=False, dtype=tf.float32)
        img = tf.image.resize(img, target_size)
        img = tf.keras.applications.mobilenet_v2.preprocess_input(img * 255.0)
        img = tf.expand_dims(img, axis=0)
        return img
    except Exception as e:
        print(f"Error loading or processing image {image_path}: {e}")
        return None

# (Prediction and metric discussion part remains the same)
if training_successful:
    image_path_input = input("Enter the full path to the input image: ")
    if not os.path.exists(image_path_input):
        print(f"Error: File '{image_path_input}' not found.")
    else:
        input_image = load_and_preprocess_image(image_path_input)
        if input_image is not None:
            print(f"\nMaking prediction for image: {image_path_input}")
            start_time = time.time()
            predictions = model.predict(input_image)
            predicted_mask = np.argmax(predictions, axis=-1)[0]
            end_time = time.time()
            print(f"Prediction completed in {end_time - start_time:.2f} seconds.")

            # (Visualization remains the same)
            plt.figure(figsize=(12, 6))
            plt.subplot(1, 2, 1); plt.title("Input Image (preprocessed)")
            display_img = (input_image[0].numpy() + 1.0) / 2.0
            plt.imshow(np.clip(display_img, 0, 1)); plt.axis('off')
            plt.subplot(1, 2, 2); plt.title("Predicted Segmentation Mask")
            plt.imshow(predicted_mask, cmap='jet', vmin=0, vmax=NUM_CLASSES-1); plt.axis('off')
            plt.colorbar(label='Predicted Class'); plt.show()

            # (Metric Discussion remains the same)
            print("\n--- Performance Evaluation ---")
            print("IMPORTANT: To calculate objective performance metrics for this specific prediction,")
            print("we would need the TRUE (ground truth) full segmentation mask corresponding to this input image.")
            # ... (rest of metric discussion) ...
            print("\nCommon Metrics for Semantic Segmentation:") # ... (details remain same) ...
            print("\n--- Performance Target (95%) ---") # ... (details remain same) ...
            print("\nThe current model state (after brief dummy training with regularization) might still have low performance on real data.")
            print("This code serves as a STARTING POINT including regularization techniques.")
```
