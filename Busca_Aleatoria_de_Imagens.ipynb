```python
import os
import shutil
from bing_image_downloader import downloader

# Função para baixar imagens
def download_images(query, limit):
    downloader.download(query, limit=limit, output_dir='dataset', adult_filter_off=True, force_replace=False, timeout=60)

# Função para criar um arquivo ZIP com as imagens baixadas
def zip_folder(folder_name, output_name):
    shutil.make_archive(output_name, 'zip', folder_name)

# Lista de animais
animals = ['Cachorro', 'Coelho', 'Gato', 'Hamster', 'Passarinho', 'Peixe', 'Leão', 'Girafa', 'Zebra']

# Diretório de download
output_dir = 'dataset'

# Download das imagens
for animal in animals:
    print(f"Baixando imagens de {animal}...")
    download_images(animal, limit=20)

# Criar arquivo ZIP com todas as imagens baixadas
zip_folder(output_dir, 'images_dataset')
print("Download concluído! Imagens foram salvas no arquivo 'images_dataset.zip'.")
```
